#轉年月份

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#匯入
scaler = StandardScaler()
SO2 = pd.read_csv("/content/gdrive/MyDrive/DM/Score/library_in_and_out_record.csv", encoding='big5') #unicode_escape
SO2.drop(SO2[SO2['管制項目']== '出' ].index,inplace=True)
datedeal=pd.to_datetime(SO2['刷卡時間'])
#提取年月
y=pd.DataFrame(datedeal.dt.year)
m=pd.DataFrame(datedeal.dt.month)
# 轉年
y=y-1911
print(y)
# 轉上下學期
Change = {1: 2, 2:2 ,3: 1,4: 1,5: 1,6: 1,7: 1,8: 1,9:2,10:2,11:2,12:2}
m['刷卡時間'] = m['刷卡時間'].map(Change)
print(m)
# 合併年份學期
D = y.merge(m, how='inner', left_index=True, right_index=True,suffixes=['_年','_月'])
DD = SO2.merge(D, how='inner', left_index=True, right_index=True)
DD=DD.drop(['刷卡時間'],axis=1)
# 整合
A = pd.DataFrame(DD.pivot_table(columns=["學生","管制項目","刷卡時間_年","刷卡時間_月"], aggfunc='size'))
# 輸出csv
A.to_csv("/content/gdrive/MyDrive/DM/Score/count_in_out.csv" ,index=False,encoding='big5')
print("ok")
```

#轉年月份(借閱)

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#
scaler = StandardScaler()
SO2 = pd.read_csv("/content/gdrive/MyDrive/DM/Score/borrowing_record.csv", encoding='big5') #unicode_escape
datedeal=pd.to_datetime(SO2['借書時間'])
#提取年月
y=pd.DataFrame(datedeal.dt.year)
m=pd.DataFrame(datedeal.dt.month)
# 轉年
y=y-1911
# 轉上下學期
Change = {1: 2, 2:2 ,3: 1,4: 1,5: 1,6: 1,7: 1,8: 1,9:2,10:2,11:2,12:2}
m['借書時間'] = m['借書時間'].map(Change)
# 合併
D = y.merge(m, how='inner', left_index=True, right_index=True,suffixes=['_年','_月'])
DD = SO2.merge(D, how='inner', left_index=True, right_index=True)
DD=DD.drop(['借書時間'],axis=1)
#整合
A = pd.DataFrame(DD.pivot_table(columns=["學生","借書時間_年","借書時間_月"], aggfunc='size'))
# 輸出csv
# DD.to_csv("/content/gdrive/MyDrive/DM/Score/bor.csv" ,index=False,encoding='big5')
A.to_csv("/content/gdrive/MyDrive/DM/Score/count_bor.csv" ,index=False,encoding='big5')
print("ok")
```

#合併

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#匯入
inout = pd.read_csv("/content/gdrive/MyDrive/DM/Score/in_out.csv", encoding='big5') #unicode_escape
bor = pd.read_csv("/content/gdrive/MyDrive/DM/Score/bor.csv", encoding='big5') #unicode_escape
traincl = pd.read_csv("/content/gdrive/MyDrive/DM/Score/course_train_cl.csv", encoding='big5') #unicode_escape
#刪除108 100
inout.drop(inout[inout['學年']== 108 ].index,inplace=True)
inout.drop(inout[inout['學年']== 100 ].index,inplace=True)
inout=inout.drop(['管制項目'],axis=1)
bor.drop(bor[bor['學年'] == 108].index,inplace=True)
bor.drop(bor[bor['學年'] == 100].index,inplace=True)
traincl=traincl.drop(['借書次數','進'],axis=1)
traincl.drop(traincl[traincl['學年'] == 100].index ,inplace=True)
# 合併
N = bor.merge(inout, on=['學生','學年','學期'],how='outer')#left_index=True, right_index=True
NN=traincl.merge(N, on=['學生','學年','學期'],how='outer')
# nan補0、改名
NN = NN.fillna(0) #補0
NN.rename(columns={'count_x': '借閱次數','count_y': '進出次數'}, inplace=True)
# 輸出csv
NN.to_csv("/content/gdrive/MyDrive/DM/Score/N.csv" ,index=False,encoding='big5')
```

```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#匯入
train_107 = pd.read_csv("/content/gdrive/MyDrive/DM/Score/107-2_train.csv", encoding='big5') #unicode_escape
test_107 = pd.read_csv("/content/gdrive/MyDrive/DM/Score/107-2_test.csv", encoding='big5') #unicode_escape
# 刪除特定欄
train_107=train_107.drop(['學生','學年','學期','概化排名'],axis=1)
test_107=test_107.drop(['學生','學年','學期','概化排名','排名差'],axis=1)
# 正規訓練
scaler.fit(train_107) 
train_features = scaler.transform(train_107)
w= pd.DataFrame(train_features)
X=w.loc[:,:1]
y=train_107["排名差"]
# 正規測試
scaler.fit(test_107)
train_features = scaler.transform(test_107)
t= pd.DataFrame(train_features)
Xt=t.loc[:,:1]
# 分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
# 模型
Train=RandomForestRegressor(n_estimators=20, criterion = 'absolute_error',random_state=0)
train = Train.fit(X_train, y_train)
Pred =train.predict(X_test)
Pred1=pd.DataFrame(Pred)
Ans=pd.DataFrame(y_test)  
pred =train.predict(Xt)
print(pred)
# 匯入csv
w=[]
for x in range(0,Xt.shape[0]):
  w.append(pred[x])
w_rf=pd.DataFrame(w,columns=['排名差'])
w_rf.to_csv("/content/gdrive/MyDrive/DM/Score/ANS.csv" ,index=True)
print("ok")
# '''
```
